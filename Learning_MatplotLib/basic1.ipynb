```python
import matplotlib.pyplot as plt

month = [1,2,3,4]
sales = [100,500,300,400]
plt.plot(month,sales, color ="blue", linestyle='--', linewidth = 2, marker ='o',label = "2025 sales")
plt.xlabel("Months", color = "red")
plt.legend(loc="upper right")
plt.grid(color = "black", linewidth = 0.2)
plt.xticks([1,2,3,4],['M1',"M2",'M3','M4'])
```

```python
import matplotlib.pyplot as plt
# plt.pie(value,labels , colors,autopct)
regions = ['North','South','East','West']
revenue = [3000,2000,1100,4120]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title("Revenue per capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
scores = [45,34,53,91,12,65,23,11,65,76,32,87,17]

plt.hist(scores,bins=5,color = 'orange',edgecolor ='black')
plt.title('Student scores')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# plt.scatter(x,y,color,marker,labelname)
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies,exam_scores,color = "green",label ='student data',marker='o')
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.title("Prediction of scores on hours studied")
plt.grid(color = 'green')
plt.legend()
plt.show()
```

```python
# Generating two scatter plots in same canvas

import matplotlib.pyplot as plt

# plt.scatter(x,y,color,marker,labelname)

plt.scatter([1,2,3],[50,60,70],color = "green",label ='classA',marker='o')
plt.scatter([1,2,3],[55,80,90],color = "blue",label ='classB',marker='o')
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.title("Prediction of scores on hours studied")
plt.grid(color = 'green')
plt.legend()
plt.show()
```
