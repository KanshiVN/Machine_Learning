```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,14,21]

plt.subplot(1,3,1) #1row, 2 col, 1st subplot  -. THIS IS THE INITIAL THIS TELLS (HOW MANY ROWS, HOWMANY COLUMS, CURRENT INDEX)
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,3,2) #1row,2 col, 2nd subplot
plt.bar(x,y)
plt.title("Bar chart")

plt.subplot(1,3,3) #1row,3 col, 3nd subplot
plt.bar(x,y)
plt.title("Bar chart")

# plt.tight_layout() -> fits the whole canvas

plt.show()
```

```python
# object oriented
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,2,figsize = (10,5))
x = [1,2,3,4]
y = [10,20,14,21]

ax[0].plot(x,y)
ax[0].set_title('line plot')

ax[1].bar(x,y)
ax[1].set_title('Bar plot')

plt.show()
```
